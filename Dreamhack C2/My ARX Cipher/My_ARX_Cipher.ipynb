```python
def decrypt_4bytes(third_v3, third_v5, key_list):
    """4바이트 암호화 데이터를 받아 원래의 4바이트(v3, v4)로 되돌리는 함수"""
    curr_v3 = third_v3
    curr_v5 = third_v5
    
    for i in range(2, -1, -1):
        x3 = key_list[2*i+1] ^ curr_v3
        v4 = (x3 >> 7) | ((x3 << 9) & 0xFFFF)
        x5 = key_list[2*i] ^ curr_v5
        y5 = (x5 - v4) & 0xFFFF
        v3 = (y5 >> 7) | ((y5 << 9) & 0xFFFF)
        
        curr_v3 = v3
        curr_v5 = v4
        
    return curr_v3, curr_v5


# 1. 키 배열 초기화 (SuP3RSaFeK3Y -> 16비트 정수 6개)
a1_str = "SuP3RSaFeK3Y"
key_list = []
for k in range(0, len(a1_str), 2):
    low = ord(a1_str[k])
    high = ord(a1_str[k+1])
    key_list.append((high << 8) | low)

# 2. 파일 복호화 루프 실행
# "encrypted.enc" 파일 위치에 실제 복호화할 파일 경로를 입력하세요.
enc_file_path = "flag.enc"  
dec_file_path = "decrypted.txt"

with open(enc_file_path, "rb") as f_in, open(dec_file_path, "wb") as f_out:
    while True:
        # 파일에서 4바이트씩 chunk를 읽어옴
        chunk = f_in.read(4)
        
        # 파일 끝에 도달하면 반복문 종료
        if not chunk:
            break
            
        # 만약 마지막 남은 데이터가 4바이트 미만인 경우 (예외 처리)
        if len(chunk) < 4:
            f_out.write(chunk)
            break
            
        # 읽어온 4바이트를 리틀 엔디안 방식으로 2바이트씩 쪼갬
        third_v3 = int.from_bytes(chunk[0:2], byteorder='little')
        third_v5 = int.from_bytes(chunk[2:4], byteorder='little')
        
        # 역산 함수 실행
        orig_v3, orig_v4 = decrypt_4bytes(third_v3, third_v5, key_list)
        
        # 구한 복호화 데이터(정수)를 다시 리틀 엔디안 바이너리로 변환
        dec_bytes = orig_v3.to_bytes(2, byteorder='little') + orig_v4.to_bytes(2, byteorder='little')
        
        # 결과 파일에 순서대로 쓰기
        f_out.write(dec_bytes)

```
