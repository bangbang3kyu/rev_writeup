```python
hex_data = "220c6a33204455fb390074013c4156d704316528205156d70b217c14255b6ce10837651234464e"
print(type(bytes.fromhex(hex_data))) 
encoded_bytes = bytes.fromhex(hex_data)
key_list = [102, 68, 17, 119, 85, 34, 51, 136]
result = []

print(encoded_bytes[0])

for i in range(len(encoded_bytes)):
    result.append(encoded_bytes[i]^key_list[i%8])

print(result)

final_string = ""
for j in range(len(result)):
    final_string+=chr(result[j])

print(final_string)
```
