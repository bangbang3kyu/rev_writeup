```python
from base64 import b64encode, b64decode

with open("flag_out.txt", "rb") as f: enc_flag = f.read()
with open("text_in.txt", "rb") as f: textin = f.read()
with open("text_out.txt", "rb") as f: textout = f.read()

textb64 = b64encode(textin)
custom_table = {}
for a, b in zip(textout, textb64):
    custom_table[a] = b

print(bytes(custom_table))

flag_b64enc = b""
for c in enc_flag:
    flag_b64enc += bytes([custom_table[c]])
    print(flag_b64enc)
print(b64decode(flag_b64enc))
```
