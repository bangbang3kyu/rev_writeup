```python
arr = [82, 223, 179, 96, 241, 139, 28, 181, 87, 209, 159, 56,75, 41, 217, 38, 127, 201, 163, 233, 83, 24, 79, 184,106, 203, 135, 88, 91, 57, 30]

a1 = []

for i in range(0,31):
    target = i ^ arr[i]
    #print(target)
    shift = i&7
    if shift == 0:
        a1.append(target)
    else:
        a1.append(((target >> shift) & 255) | ((target << (8 - shift)) & 255))
        

#print("정수 리스트:", a1)

result = []

for i in a1:
    result.append(chr(i))

print("".join(result))
```
