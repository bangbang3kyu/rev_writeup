```python
arr = [82, 223, 179, 96, 241, 139, 28, 181, 87, 209, 159, 56,75, 41, 217, 38, 127, 201, 163, 233, 83, 24, 79, 184,106, 203, 135, 88, 91, 57, 30]

a1 = []

for i in range(0,31):
    target = i ^ arr[i]
    # i ^ __ROL1__(입력값[i], 회전수) 이거니까 역산을 위해 i값을 xor 후 ROL1 역산을 진행
    # print(target)
    shift = i&7 # 회전수
    if shift == 0: # 회전수 0이면 그냥 그대로니까 if문 설정
        a1.append(target)
    else:
        a1.append(((target >> shift) & 255) | ((target << (8 - shift)) & 255))
        

# print(a1)

result = []

for i in a1:
    result.append(chr(i))

print("".join(result))
```
