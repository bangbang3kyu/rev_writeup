```python
enc_file_path = "secretMessage.enc"
dec_file_path = "secretMessage.raw"

with open(enc_file_path, "rb") as f_in, open(dec_file_path, "wb") as f_out:
    data = f_in.read()
    i = 0
    n = len(data)

    while i < n:
        if i + 2 < n and data[i] == data[i + 1]:
            byte_val = data[i]
            count = data[i + 2]

            # 앞의 같은 바이트 2개 + 추가 반복 count개
            f_out.write(bytes([byte_val]) * (2 + count))

            i += 3
        else:
            f_out.write(bytes([data[i]]))
            i += 1

print("복호화 완료!")
```

```python
from PIL import Image

enc_file_path = "secretMessage.enc"
raw_file_path = "secretMessage.raw"
png_file_path = "secretMessage.png"

with open(enc_file_path, "rb") as f_in, open(raw_file_path, "wb") as f_out:
    data = f_in.read()
    i = 0
    n = len(data)

    while i < n:
        if i + 2 < n and data[i] == data[i + 1]:
            byte_val = data[i]
            count = data[i + 2]

            # 앞의 같은 바이트 2개 + 추가 반복 count개
            f_out.write(bytes([byte_val]) * (2 + count))

            i += 3
        else:
            f_out.write(bytes([data[i]]))
            i += 1

print("복호화 완료!")

with open(raw_file_path, "rb") as f:
    output = f.read()

print("raw size:", len(output))

img = Image.frombytes("1", (500, 50), output)
img.save(png_file_path)

print("이미지 저장 완료:", png_file_path)
```
