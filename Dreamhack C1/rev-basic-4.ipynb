```python
arr = [36, 39, 19, 198,198, 19, 22, 230, 71, 245, 38, 150, 71, 245, 70, 39, 19, 38,38, 198, 86, 245, 195,195, 245, 227,227]
a1 = []

for i in range(0,len(arr)):

    val = (arr[i]) & 0xFF # 값이 나중에 255가 넘지않도록 0xFF를 "&"해주었다.
    result = (((val >> 4) & 0x0F) | ((val << 4) & 0xF0)) # 역산을 위해 니블 스왑 후 result에 저장
    a1.append(result)

print(a1)

# print(chr(97))
a1_result = []

for i in a1:
    a1_result.append(chr(i))

print("".join(a1_result))
```
